```python
import pandas as pd

# File path for the CSV file
csv_file_path = 'insights_from_CATEGORICAL_FORMATTED_VIKASPEDIA.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)

# Select and display only the first three columns
df = df.iloc[:, :3]  # Selecting the first three columns
df
```

```python
df = df.sort_values(by='TOTAL_TOKENS(excluding_spaces)')
df.reset_index(inplace=True, drop=True)
df
```

```python

# Count the number of values less than 4050
equal_to_0 = (df['TOTAL_TOKENS(excluding_spaces)'] ==0 ).sum()

# Count the number of values less than 4050
less_than_4050 = (df['TOTAL_TOKENS(excluding_spaces)'] <= 4050).sum()

# Count the number of values greater than 4050
more_than_4050 = (df['TOTAL_TOKENS(excluding_spaces)'] > 4050).sum()

print(f"Values equal to 0: {equal_to_0}")
print(f"Values less than 4050: {less_than_4050}")
print(f"Values more than 4050: {more_than_4050}")
```

```python
import os

def copy_files_based_on_condition(df, source_dir, dest_dir, categories, condition):
    filtered_data = df[condition]
    for index, row in filtered_data.iterrows():
        source_file_path = os.path.join(source_dir, row['CATEGORY'], row['FILE_NAME'])
        with open(source_file_path, "r", encoding="utf-8") as source_file:
            dest_file_path = os.path.join(dest_dir, categories, row['CATEGORY'], row['FILE_NAME'])
            os.makedirs(os.path.join(dest_dir, categories, row['CATEGORY']), exist_ok=True)
            with open(dest_file_path, "w", encoding="utf-8") as dest_file:
                dest_file.write(source_file.read())

source = "CATEGORICAL_FORMATTED_VIKASPEDIA"
dest = "SPLITTED_CATEGORICAL_FORMATTED_VIKASPEDIA"
categories = ["equal_to_0", "less_than_4050", "more_than_4050"]

equal_to_0_condition = df['TOTAL_TOKENS(excluding_spaces)'] == 0
less_than_4050_condition = df['TOTAL_TOKENS(excluding_spaces)'] <= 4050
more_than_4050_condition = df['TOTAL_TOKENS(excluding_spaces)'] > 4050

copy_files_based_on_condition(df, source, dest, categories[0], equal_to_0_condition)
copy_files_based_on_condition(df, source, dest, categories[1], less_than_4050_condition)
copy_files_based_on_condition(df, source, dest, categories[2], more_than_4050_condition)
```
