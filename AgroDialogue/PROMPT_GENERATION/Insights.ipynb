```python
import os
import nltk
import re
```

```python
# Function to calculate statistics for a given text
def calculate_statistics(text):
    tokens = nltk.word_tokenize(text)
    # Exclude tokens that are just spaces
    tokens_except_spaces = [token for token in tokens if token.strip()]
    total_tokens = len(tokens_except_spaces)
    total_sentences = len(nltk.sent_tokenize(text))
    total_words = len(tokens)
    unique_words = set(tokens)
    return total_tokens, total_sentences, total_words, unique_words
```

```python
def giveInsights(DIR):
    total_tokens = 0
    total_sentences = 0
    total_words = 0
    total_lines=0
    unique_words = []
    if os.path.isdir(DIR):
        ALL_FILENAMES=os.listdir(DIR)
        for filename in ALL_FILENAMES:
            item_path = os.path.join(DIR, filename)
            with open(item_path, "r", encoding="utf-8") as fil:
                file_content = fil.read()
                file_content = file_content.strip()
                tokens, sentences, words, unique = calculate_statistics(file_content)
                total_tokens += tokens
                total_sentences += sentences
                total_lines += file_content.count('\n') + 1
                total_words += words
                unique_words+=unique
    else:
        with open(DIR, "r", encoding="utf-8") as fil:
                file_content = fil.read()
                file_content = file_content.strip()
                tokens, sentences, words, unique = calculate_statistics(file_content)
                total_tokens += tokens
                total_sentences += sentences
                total_lines += file_content.count('\n') + 1
                total_words += words
                unique_words+=unique
        ALL_FILENAMES=[DIR]
    return len(ALL_FILENAMES),total_tokens,total_sentences,total_lines,total_words,len(set(unique_words))
```

```python
def showInsights(DIR):
    allData=giveInsights(DIR)
    print(f"TOTAL FILES: {allData[0]}")
    print(f"TOTAL TOKENS (excluding spaces): {allData[1]}")
    print(f"TOTAL SENTENCES: {allData[2]}")
    print(f"TOTAL LINES: {allData[3]}")
    print(f"TOTAL WORDS: {allData[4]}")
    print(f"UNIQUE WORDS: {allData[5]}")
```

```python
print("ALL FILES")
showInsights("FILEBASED/ALL")
print("\n\nPLANTIX")
showInsights("FILEBASED/PLANTIX")
print("\n\nVIKASPEDIA")
showInsights("FILEBASED/NEWVIKASPEDIA")
```

```python
import pandas as pd
dict={}
for item in os.listdir("CATEGORICAL_FORMATTED_VIKASPEDIA"):
    allData=giveInsights(os.path.join("CATEGORICAL_FORMATTED_VIKASPEDIA",item))
    dict[item]={
    "totalFiles":allData[0],
    "totalTokens":allData[1],
    "totalSentences":allData[2],
    "totalLines":allData[3],
    "totalWords":allData[4],
    "uniqueWords":allData[5]
    }
df=pd.DataFrame(dict).transpose()
df.loc['Column_Total'] = df.sum()
df
```

```python
import pandas as pd

data_dict = {}
for category in os.listdir("CATEGORICAL_FORMATTED_VIKASPEDIA"):
    for file in os.listdir(os.path.join("CATEGORICAL_FORMATTED_VIKASPEDIA", category)):
        file_path = os.path.join("CATEGORICAL_FORMATTED_VIKASPEDIA", category, file)
        allData = giveInsights(file_path)
        data_dict[file] = {
            "CATEGORY": category,
            "TOTAL_TOKENS(excluding_spaces)": allData[1],
            "TOTAL_SENTENCES": allData[2],
            "TOTAL_LINES": allData[3],
            "TOTAL_WORDS": allData[4],
            "UNIQUE_WORDS": allData[5]
        }

# Create a DataFrame directly from the dictionary
df = pd.DataFrame(data_dict).transpose()

# Rename the index name to 'FILE_NAME'
df.index.name = 'FILE_NAME'

df
```

```python
csv_file_path = 'insights_from_CATEGORICAL_FORMATTED_VIKASPEDIA.csv'
df.to_csv(csv_file_path)
print(f"DataFrame saved to '{csv_file_path}'")
```
