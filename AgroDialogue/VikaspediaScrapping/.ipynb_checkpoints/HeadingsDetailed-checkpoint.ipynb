```python
import requests
```

```python
url="https://vikaspedia.in/agriculture"
```

```python
dat=requests.get(url,verify=False)
print(dat.status_code)
# Informational responses (100 – 199)
# Successful responses (200 – 299)
# Redirection messages (300 – 399)
# Client error responses (400 – 499)
# Server error responses (500 – 599)
```

```python
from bs4 import BeautifulSoup
soup=BeautifulSoup(dat.text,"html.parser")
print(soup)
```

```python
a_tag = soup.find_all('a', class_='folderfile_name')
a_tag
```

```python
folderfile_name=[]
for i in a_tag:
    folderfile_name.append(i.text)
folderfile_name
```

```python
with open('folderCategories.txt', 'w') as file:
    for item in folderfile_name:
        file.write(f"{item}\n")
```

```python
#encode urls like url for agri-credit
def encodeURLS(url,link):
    link=link.lower()
    link=link.replace(" ","-")
    return url+"/"+link
```

```python
for i in range(0,len(folderfile_name)):
    folderfile_name[i]=encodeURLS(url,folderfile_name[i])
folderfile_name
```

```python
from bs4 import BeautifulSoup
def fetchURL(URL):
    dat=requests.get(URL,verify=False)
    soup=BeautifulSoup(dat.text,"html.parser")
    return soup
```

```python
newsoup=fetchURL(folderfile_name[0])
a_tag = newsoup.find_all('a', class_='folderfile_name')
a_tag
```
