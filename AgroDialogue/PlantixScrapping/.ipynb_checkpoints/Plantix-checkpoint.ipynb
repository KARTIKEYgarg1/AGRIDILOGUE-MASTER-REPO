```python
import requests
from bs4 import BeautifulSoup
import selenium
```

```python
with open("html.txt",encoding="utf-8") as file:
    htmlcontent=file.read()
```

```python
# parse the content
soup = BeautifulSoup(htmlcontent,'html.parser')
# print(soup.prettify)
```

```python
title= soup.title
print(title.text)
```

```python
a_tags = soup.find_all('a',class_="disease-content")
print(a_tags)
```

```python
diseaseContent = soup.find_all('a',class_="disease-content")
len(diseaseContent)
```

```python
disease_names = soup.find_all('h3',class_="disease-name textfade")
len(disease_names)
```

```python
diseaseNamesList=[]
for i in disease_names:
    diseaseNamesList.append(i.text)
```

```python
diseaseType = soup.find_all('p',class_="disease-type textfade")
len(diseaseType)
```

```python
diseaseTypeList=[]
for i in diseaseType:
    diseaseTypeList.append(i.text)
```

```python
list(set(diseaseTypeList)),len(list(set(diseaseTypeList)))
```

```python
resultCard = soup.find_all('div',class_="pests-and-diseases-result result-card")
links=[]
images=[]
for i in resultCard:
    links.append(i.a.get("href"))
    temp=i.find("img")
    if temp:
        images.append(temp.get("src"))
    else:
        images.append("NA")
```

```python
images.count("NA") #414 missing images
```

```python
resultingDictionary={}
for i in range(700):
    resultingDictionary[diseaseNamesList[i].strip()]={"id":"DN"+(str)(i+1),"diseaseType":diseaseTypeList[i],"links":"https://plantix.net/en/library/"+links[i],"images":images[i]}
```
