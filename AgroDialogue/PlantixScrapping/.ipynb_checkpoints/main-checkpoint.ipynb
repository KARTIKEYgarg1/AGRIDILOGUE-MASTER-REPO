```python
import requests
from bs4 import BeautifulSoup
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import os
path="C:\Program Files (x86)\msedgedriver.exe"
```

```python
def load_all_diseases():
    with open("html.txt",encoding="utf-8") as file:
        diseaseNamesList=[]
        diseaseTypeList=[]
        links=[]
        images=[]
        resultingDictionary={}
        htmlcontent=file.read() #using downloaded html content
        soup = BeautifulSoup(htmlcontent,'html.parser')
        disease_names = soup.find_all('h3',class_="disease-name textfade")
        diseaseType = soup.find_all('p',class_="disease-type textfade")
        resultCard = soup.find_all('div',class_="pests-and-diseases-result result-card")
        for i in disease_names:
            diseaseNamesList.append(i.text)
        for i in diseaseType:
            diseaseTypeList.append(i.text)
        for i in resultCard:
            links.append(i.a.get("href"))
            temp=i.find("img")
            if temp:
                images.append(temp.get("src"))
            else:
                images.append("NA")
        for i in range(700):
            resultingDictionary[diseaseNamesList[i].strip()]={"id":"DN"+(str)(i+1),"diseaseType":diseaseTypeList[i],"link":"https://plantix.net/en/library/"+links[i],"images":images[i]}
        return resultingDictionary
```

```python
result=load_all_diseases()
result
```

```python
def load_disease(url):
    try:
        str=""
        driver = webdriver.Edge() 
        driver.get(url)
        title=driver.find_element(By.TAG_NAME, 'h1').text
        str=str+title
        symptoms = driver.find_element(By.CLASS_NAME, "symptoms")
        str=str+"\n\n"+symptoms.text
        str=str+"\n\nProduct Recommendations\n"
        recommendations = driver.find_element(By.CLASS_NAME, "product-recommendations")
        str=str+recommendations.text
        triggerCard = driver.find_element(By.CLASS_NAME, "trigger-card")
        str=str+"\n\n"+triggerCard.text
        str=str+"\n\nPreventive Measures \n"
        preventiveMeasures = driver.find_element(By.CLASS_NAME, "preventive-measures-list")
        str=str+preventiveMeasures.text
        time.sleep(5)
        driver.quit() 
        with open("./PlantixData/" + title + ".txt", "w", encoding="utf-8") as file:
            file.write(str)
        print("Success for "+title)
        return str
    except Exception as e:
        print("EXCEPTION OCCURRED",e)
        return "NA"
```
